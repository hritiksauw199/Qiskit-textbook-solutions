```python
from qiskit import QuantumCircuit, assemble, Aer, execute
from math import pi, sqrt
from qiskit.visualization import plot_bloch_multivector, plot_histogram
import numpy as np
```

# 1. What would be the resulting state of the control qubit (q0) if the target qubit (q1) was in the state |0⟩? (as shown in the circuit below)? Use Qiskit to check your answer.

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.draw()
```

```python
sim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
res = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(res)
```

**Applying Controlled-T gate**

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.cp(pi/4, 0, 1)
qc.draw()
```

```python
sim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
res = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(res)
```

# 2. What would happen to the control qubit (q0) if the target qubit (q1) was in the state |1⟩, and the circuit used a controlled-Sdg gate instead of the controlled-T (as shown in the circuit below)?

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.draw()
```

```python
sim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
res = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(res)
```

**Applying controlled-Sdg**

```python
qc = QuantumCircuit(2)
qc.h(0)
qc.x(1)
qc.cp(-pi/2, 0, 1)
qc.draw()
```

```python
sim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
res = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(res)
```

# 3. What would happen to the control qubit (q0) if it was in the state |1⟩ instead of the state |+⟩ before application of the controlled-T (as shown in the circuit below)?

```python
qc = QuantumCircuit(2)
for i in range(2):
    qc.x(i)
qc.draw()
```

```python
sim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
res = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(res)
```

**After application of Controlled-T gate**

```python
qc = QuantumCircuit(2)
for i in range(2):
    qc.x(i)
qc.cp(pi/4, 0, 1)
qc.draw()
```

```python
sim = Aer.get_backend('aer_simulator')
qc.save_statevector()
qobj = assemble(qc)
res = sim.run(qobj).result().get_statevector()
plot_bloch_multivector(res)
```
